```python
def clean(x):
    if "M" in x:
        return float(x[:-1])*1024
```

```python
df.Size = df["Size"].apply(clean)
```

```python
df.Installs.dropna(inplace=True)
```

```python
def clean_1(x):
    if "+" in x:
        return float(x[:-1].replace(",",""))
```

```python
df["Installs"]=df["Installs"].apply(clean_1)
```

```python
def clean_2(x):
    if "and up" in str(x):
        return x[:-6]
```

```python
df["Android Ver"]=df["Android Ver"].apply(clean_2)
```

```python
def detect_out(x):
    out=[]
    Q1=x.quantile(.25)
    Q3=x.quantile(.75)
    IQR=Q3-Q1
    upper=Q3+(1.5*IQR)
    lower=Q1-(1.5*IQR)
    for i in x:
        if i>upper or i < lower:
            out.append(i)
    return out
```
